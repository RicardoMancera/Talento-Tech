Load the diabetes dataset, divided into `X` data and `y` features (as arrays):

````python
(X, y) = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
````

Select just one feature to target for this exercise

````python
# Selecting the 3rd feature
X = X[:, 2]
print(X.shape)
print(X[0])
X
````

````python
#Reshaping to get a 2D array
X = X.reshape(-1, 1)
print(X.shape)
print(X)
````

Split the training and test data for both `X` and `y`

````python
(X_train, X_test, y_train, y_test) = model_selection.train_test_split(X, y, test_size=0.33)
````

Select the model and fit it with the training data

````python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
````

Use test data to predict a line

````python
y_pred = model.predict(X_test)
````

Display the results in a plot

````python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
````

````python
import pandas as pd
import seaborn as sns
````

gráfico de distribución del error

````python
error = y_test - y_pred

plt.figure(figsize=(3,3))
sns.histplot(error, bins=30, kde=True, color='blue')
plt.title('Distribución del Error', fontsize=8)
plt.xlabel('Error', fontsize=8)
plt.ylabel('Frecuencia', fontsize=8)
plt.show()
````

Error cuadrado

````python
errorCuadratico = (y_test - y_pred)**0.5
plt.figure(figsize=(3, 3))
sns.histplot(errorCuadratico, bins=30, kde=True, color='blue')
plt.title('Distribución del Error cuadrático', fontsize=8)
plt.xlabel('Error', fontsize=8)
plt.ylabel('Frecuencia', fontsize=8)
plt.show()
````

````python
from sklearn.metrics import mean_squared_error
````

````python
mse = mean_squared_error(y_test, y_pred)
print("Error cuadratico medio MSE:", mse)
````

Error cuadratico medio MSE (Manual)

````python
def MSE (y_test, y_pred):
 MSE = np.mean((y_test - y_pred)**2)
 return MSE
print("Error cuadratico medio MSE manual:", MSE(y_test, y_pred))
````

Raíz del error cuadratico medio RMSE

````python
rmse = np.sqrt(mse) # sqrt = Raiz Cuadrada
print("Raíz del error cuadratico medio RMSE:", rmse)
````

````python
def RMSE (y_test, y_pred):
  RMSE = (np.mean((y_test - y_pred)**2))**0.5
  return RMSE
print("Raíz del error cuadratico medio RMSE manual:", RMSE(y_test, y_pred))
````

````python
from sklearn.metrics import mean_absolute_error
````

Promedio de errores absolutos

````python
mae = mean_absolute_error(y_test, y_pred)
print('MAE:', mae)
````

Cálculo del promedio de errores absolutos manual

````python
def MAE (y_test, y_pred):
  MAE = np.mean(np.abs(y_test - y_pred)) # abs = absoluto
  return MAE
print('MAE manual:', MAE(y_test, y_pred))
````

Cálculo del error de porcentaje medio absoluto

````python
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print('MAPE:',mape,'%')
````

````python
from sklearn.metrics import r2_score
````

````python
Coeficiente de determinación
````

````python
r2 = r2_score(y_test, y_pred)
print('R2:', r2)
````

````python
def R2 (y_test, y_pred):
  R2 = 1 - (np.sum((y_test - y_pred)**2) / np.sum((y_test - np.mean(y_test))**2))
  return R2
print('R2 manual:', R2(y_test,y_pred))
````
